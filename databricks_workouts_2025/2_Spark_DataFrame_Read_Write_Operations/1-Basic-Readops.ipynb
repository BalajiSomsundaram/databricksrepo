```python
#1. Using simple string format of define custom simple schema.
own_str_struct ="iid int, ffname string, lname string, age integer, prof string"
csv_df10 = spark.read.schema(own_str_struct).csv("/Volumes/catalog1_dropme/schema1_dropme/volume1_dropme/directory_dropme/custs")
print(csv_df10.printSchema())
csv_df10.show(3)
```

```python
from pyspark.sql.types import StructType,StructField,StringType,IntegerType
custom_schema = StructType([StructField("id",IntegerType(),False),StructField("fname",StringType(),True),StructField("lname",StringType(),True),StructField("age",IntegerType(),False),StructField("prof",StringType(),True)])
csv_cust_schema_df1 = spark.read.schema(custom_schema).csv("/Volumes/catalog1_dropme/schema1_dropme/volume1_dropme/directory_dropme/custs")
print(csv_cust_schema_df1.printSchema())
csv_cust_schema_df1.show(3) 
```
